```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
```

```python
Data_PATH = 'C:/Users/Владислав/housing-price-prediction/data/raw'
df = pd.read_csv(f'{Data_PATH}/train.csv')
```

```python
df = df.drop(columns=['Alley', 'PoolQC', 'MiscFeature', 'Fence'])
```

```python
df['FireplaceQu'] = df['FireplaceQu'].fillna('NoFireplace')
```

```python
bsmt_colmns = [
    'BsmtQual', 'BsmtCond', 'BsmtExposure',
    'BsmtFinType1', 'BsmtFinType2'
]
```

```python
for colmns in bsmt_colmns:
    df[colmns] = df[colmns].fillna('NoBasement') 
```

```python
garage_colmns = [
    'GarageType', 'GarageFinish',
    'GarageQual', 'GarageCond' 
]
```

```python
for colmns in garage_colmns:
    df[colmns] = df[colmns].fillna('NoGarage') 
```

```python
df['GarageYrBlt'] = df['GarageYrBlt'].fillna(0)
```

```python
df['MasVnrArea'] = df['MasVnrArea'].fillna(0)
df['MasVnrType'] = df['MasVnrType'].fillna('None')
```

```python
df['Electrical'] = df['Electrical'].fillna(
    df['Electrical'].mode()[0]
) 
```

```python
df['LotFrontage'] = (
    df
    .groupby('Neighborhood')['LotFrontage']
    .transform(lambda x: x.fillna(x.median()))
)
```

```python
df = df.drop(columns=['Id'])
```
